# SF Salaries Exercise 

Welcome to a quick exercise for you to practice your pandas skills! We will be using the [SF Salaries Dataset](https://www.kaggle.com/kaggle/sf-salaries) from Kaggle! Just follow along and complete the tasks outlined in bold below. The tasks will get harder and harder as you go along.

** Import pandas as pd.**

```python
import pandas as pd
```

** Read Salaries.csv as a dataframe called sal.**

```python
sal = pd.read_csv('Salaries(exercise).csv')
```

** Check the head of the DataFrame. **

```python
sal.head()
```

**What is the average BasePay ?**

```python
avg_basepay = sal['BasePay'].mean()
print("The average BasePay is:", avg_basepay)
```

** What is the highest amount of OvertimePay in the dataset ? **

```python
max_overtimepay = sal['OvertimePay'].max()
print("The highest amount of OvertimePay is:", max_overtimepay)
```

** What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll). **

```python
joseph = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']
job_title = joseph['JobTitle'].iloc[0]
print("The job title of JOSEPH DRISCOLL is:", job_title)
```

** How much does JOSEPH DRISCOLL make (including benefits)? **

```python
joseph = sal[sal['EmployeeName'] == 'JOSEPH DRISCOLL']
total_pay_benefits = joseph['TotalPayBenefits'].iloc[0]
print("JOSEPH DRISCOLL makes:", total_pay_benefits)
```

** What is the name of highest paid person (including benefits)?**

```python
highest_paid_index = sal['TotalPayBenefits'].idxmax()
highest_paid_name = sal.loc[highest_paid_index, 'EmployeeName']
print("The name of the highest paid person is:", highest_paid_name)
```

** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
lowest_paid_index = sal['TotalPayBenefits'].idxmin()
lowest_paid_name = sal.loc[lowest_paid_index, 'EmployeeName']
print("The name of the lowest paid person is:", lowest_paid_name)
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
sal = sal[(sal['Year'] >= 2011) & (sal['Year'] <= 2014)]
average_basepay_by_year = sal.groupby('Year')['BasePay'].mean()
print("The average BasePay per year (2011-2014) is:\n", average_basepay_by_year)
```

** How many unique job titles are there? **

Read about the "nunique" function

```python
num_unique_job_titles = sal['JobTitle'].nunique()
print("The number of unique job titles is:", num_unique_job_titles)
```

** What are the top 5 most common jobs? **

```python
top_5_jobs = sal['JobTitle'].value_counts().head(5)
print("The top 5 most common jobs are:\n", top_5_jobs)
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
sal_2013 = sal[sal['Year'] == 2013]
job_title_counts = sal_2013['JobTitle'].value_counts()
num_job_titles_with_one_person = (job_title_counts == 1).sum()
print("The number of job titles represented by only one person in 2013 is:", num_job_titles_with_one_person)
```

** How many people have the word Chief 

---

in their job title? (This is pretty tricky) **

```python
num_chiefs = sal[sal['JobTitle'].str.contains('Chief', case=False)]['JobTitle'].count()
print("The number of people who have the word 'Chief' in their job title is:", num_chiefs)
```
